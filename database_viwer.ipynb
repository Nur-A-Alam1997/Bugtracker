```python
# import database
import sqlite3
import tkinter  as tk 
from tkinter import *
from tkinter import filedialog

def opener(self,root):
    self.opener = tk.Tk()
    # self.opener.withdraw()
    self.file_path = filedialog.askopenfilename()
    return file_path

def connect():
    return sqlite3.connect(opener())


def get_all_table_names(conn, my_conn):
    my_conn.execute("SELECT name FROM sqlite_master WHERE type='table'")
    tname=[]
    for name in my_conn.execute("SELECT name FROM sqlite_master WHERE type='table'"):
        tname.append(*name)
    return tname
# grid_frame=Frame(mainframe)

def print_database(conn, my_conn, table):
#####column names
    my_conn.execute("SELECT * FROM "+""+table+"")
    names = list(map(lambda x: x[0], my_conn.description))

    my_w = tk.Tk(table)
    my_w.title(table.upper())
    i=0
    for j in names:
        my_w.columnconfigure(i, weight=1, minsize=75)
        my_w.rowconfigure(i, weight=1, minsize=25)
        
        e=Text(my_w, height=2, width=22,font="Didot 8 bold",\
               background = "white",fg="#373737",
               borderwidth = 5,
         relief="sunken",state='normal')
        
        e.grid(row=0, column=i, sticky="ew", padx=2, pady=2) 
        e.insert(END, j)
        e.configure(state='disabled')
        i+=1
    print(names)
    
##### Dtabase values 
# my_conn.execute("SELECT * FROM complain" limit 0,5")
    i=1 
    for student in my_conn:
        
        my_w.columnconfigure(i, weight=1)#minsize=25)
        my_w.rowconfigure(i, weight=1, minsize=25)
        for j in range(len(student)):
    #         e = Entry(my_w, width=10,font="Tahoma 12", fg='blue')
    #         e.config(background="#FF6026")#,disabledbackground="red")
            
            e=Text(my_w, height=2, width=22,font="Tahoma 8",\
                   background = "white",fg="#373737",
                   borderwidth = 2,
             relief="sunken",state='normal')
            
            e.grid(row=i, column=j ,sticky="ew", padx=5, pady=5) 
            e.insert(END, student[j])
            e.configure(state='disabled')
            
        i=i+1
    my_w.mainloop()

if __name__=="__main__":
    conn=connect()
    my_conn = conn.cursor()
    print(get_all_table_names(conn, my_conn))
    print_database(conn, my_conn, table="complain")

####database in LABELS

# db = my_conn.execute("SELECT * FROM complain limit 0,10").fetchall()
#grid data
# label=Label(mainframe,text="Stored Data",bg="black",fg="white",padx=5,pady=5)
# label.config(font=("Arial",18))
# label.pack(fill="x")
# 
# grid_frame=Frame(mainframe)
# for row in range(len(db)):
#     for column in range(1,(5)):
#         label=Label(grid_frame,text=db[row][column],bg="black",fg="red",padx=5,pady=5)
#         label.config(font=("Didot",8))
#         label.grid(row=row,column=column,padx=5,pady=5,sticky="nsew")
#         grid_frame.grid_columnconfigure(column,weight=1)
# 
# grid_frame.pack(fill="x")


# with my_conn:
#     my_conn.execute("SELECT * FROM complain")
#     print(my_conn.fetchall())
```
